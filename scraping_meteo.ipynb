# Create the list of dates to scrap

```python
from datetime import date, timedelta

sdate = date(2020,1,1)   # start date
edate = date(2022,9,2) 


listedate=pd.date_range(sdate,edate-timedelta(days=1))
listedate
```

```python
newlist=[]
for date in listedate:
    newlist.append(str(date).replace(" 00:00:00","").replace("-","/"))
```

```python
newlist
```

## Scrap the defined dates

```python
column_names = ['day','tempMax', 'tempMin', 'windSpeed', 'precipitation', 'humidity', 'cloudCover', 'dayDuration', 'temp_13h', 'meteo_13h']
df_idf = pd.DataFrame(columns = column_names, dtype=object)
i=0
```

```python


for date in newlist:
    r=requests.get("https://www.historique-meteo.net/france/lyonnais/"+date)
    if i%10 == 0:
        print(str(i) + " rows done")
    soup=BeautifulSoup(r.content,"html.parser")
    table=soup.find_all('table', class_='table')
    
    rows=table[0].find_all('tr')
    
#     tempMax=rows[1].find('td', class_='text-center bg-primary').text
#     tempMin=rows[2].find('td', class_='text-center bg-primary').text
#     windSpeed=rows[3].find('td', class_='text-center bg-primary').text
#     precipitation=rows[4].find('td', class_='text-center bg-primary').text
#     humidity=rows[5].find('td', class_='text-center bg-primary').text
#     cloudCover=rows[7].find('td', class_='text-center bg-primary').text
#     dayDuration=rows[8].find('td', class_='text-center bg-primary').text
    for p in rows:
        if'Température maximale' in p.get_text():
            tempMax=p.find('td', class_='text-center bg-primary').text
        if'Température minimale' in p.get_text():
            tempMin=p.find('td', class_='text-center bg-primary').text
        if'Vitesse du vent' in p.get_text():
            windSpeed=p.find('td', class_='text-center bg-primary').text
        if'Précipitations' in p.get_text():
            precipitation=p.find('td', class_='text-center bg-primary').text
        if'Humidité' in p.get_text():
            humidity=p.find('td', class_='text-center bg-primary').text
        if'Couverture nuageuse' in p.get_text():
            cloudCover=p.find('td', class_='text-center bg-primary').text
        if'Durée du jour' in p.get_text():
            dayDuration=p.find('td', class_='text-center bg-primary').text
    meteo_13h_data = soup.find_all('div', class_='meteo_13h')  
    temp_13h = meteo_13h_data[1].find('h4').text.split(' ')[2]     
    meteo_13h = meteo_13h_data[1].find('p').text 
    row=[date,tempMax, tempMin, windSpeed,precipitation, humidity, cloudCover, dayDuration, temp_13h, meteo_13h]
    df_idf.loc[i+1] = row
    i+=1
```
