```python
import requests
import pandas as pd
import json
```

```python
url = "https://data.enedis.fr/api/v2/catalog/datasets/bilan-electrique-transpose/records?sort=jour"
payload={}
headers = {}
total_count_response = requests.request("GET", url, headers=headers, data=payload)
total_count = int(total_count_response.json()["total_count"])
print(total_count)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python

# results will be appended to this list
all_items = []

response_df = pd.DataFrame()

# loop through all offset and return JSON object
for offset in range(0, total_count, 100):

    url = "https://data.enedis.fr/api/v2/catalog/datasets/bilan-electrique-transpose/records?sort=jour&limit=100&offset="+str(offset)              
    response = requests.get(url=url, headers=headers) 
    for element in json.loads(response.text)["records"]:
        response_df = response_df.append(element["record"]["fields"], ignore_index=True)
    print(offset)
```
